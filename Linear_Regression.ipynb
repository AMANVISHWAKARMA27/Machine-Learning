```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300 + 180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
d = pd.read_csv("areas.csv")
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
#adding a new column named prices
```

```python
d
```

```python
d.to_csv("predictions.csv", index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income (1).csv")
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='green', marker = '.')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2023]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
828.46507522*2023 - 1632210.7578554575
```

```python
%matplotlib inline
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Per Capita Income (US$)', fontsize = 12)
plt.scatter(df.year, df['per capita income (US$)'], color = 'green', marker = '.')
plt.plot(df.year, reg.predict(df[['year']]), color = 'blue')
```
